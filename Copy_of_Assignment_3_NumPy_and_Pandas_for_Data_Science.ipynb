```python
#Task 1.2: NumPy Array Operations
# Convert score columns to numpy arrays
math_scores = df['math score'].values
reading_scores = df['reading score'].values
writing_scores = df['writing score'].values

# Your tasks:
# 1. Calculate mean, median, std, min, max for each subject using NumPy
# 2. Find total students who scored above 80 in math
# 3. Find total students who scored below 50 in any subject
# 4. Calculate the overall average score across all three subjects

# Expected output example:
# Math - Mean: 66.1, Median: 66.0, Std: 15.2
# Students with math > 80: 132
# Students with any score < 50: 178

# my code here:
# 1. Calculate mean, median, std, min, max for each subject using NumPy
print("Math - Mean: {:.1f}, Median: {:.1f}, Std: {:.1f}, Min: {:.1f}, Max: {:.1f}".format(
    np.mean(math_scores), np.median(math_scores), np.std(math_scores), np.min(math_scores), np.max(math_scores)
))
print("Reading - Mean: {:.1f}, Median: {:.1f}, Std: {:.1f}, Min: {:.1f}, Max: {:.1f}".format(
    np.mean(reading_scores), np.median(reading_scores), np.std(reading_scores), np.min(reading_scores), np.max(reading_scores)
))
print("Writing - Mean: {:.1f}, Median: {:.1f}, Std: {:.1f}, Min: {:.1f}, Max: {:.1f}".format(
    np.mean(writing_scores), np.median(writing_scores), np.std(writing_scores), np.min(writing_scores), np.max(writing_scores)
))

# 2. Find total students who scored above 80 in math
students_above_80_math = np.sum(math_scores > 80)
print(f"\nStudents with math > 80: {students_above_80_math}")

# 3. Find total students who scored below 50 in any subject
students_below_50_any = np.sum((math_scores < 50) | (reading_scores < 50) | (writing_scores < 50))
print(f"Students with any score < 50: {students_below_50_any}")

# 4. Calculate the overall average score across all three subjects
overall_average_score = np.mean(np.concatenate((math_scores, reading_scores, writing_scores)))
print(f"\nOverall average score: {overall_average_score:.1f}")
```

```python
#Task 2.2: Create and Fix Missing Value
# Artificially create missing data for practice
import numpy as np

# Your tasks:
# 1. Make a copy of the dataset: df_practice = df.copy()
# 2. Randomly set 50 math scores to np.nan
# 3. Randomly set 30 writing scores to np.nan
# 4. Fill missing math scores with the MEDIAN math score
# 5. Fill missing writing scores with the MEAN writing score
# 6. Verify no missing values remain

# Steps to follow:
# Step 1: Create copy and show original missing values
df_practice = df.copy()
print("Original missing values:", df_practice.isnull().sum())

# Step 2: Create missing data
np.random.seed(42)  # For reproducible results
missing_math_idx = np.random.choice(df_practice.index, 50, replace=False)
missing_writing_idx = np.random.choice(df_practice.index, 30, replace=False)

df_practice.loc[missing_math_idx, 'math score'] = np.nan
df_practice.loc[missing_writing_idx, 'writing score'] = np.nan

print("\nMissing values after creating artificial missing data:")
print(df_practice.isnull().sum())

# 3. Fill missing math scores with the MEDIAN math score
median_math_score = df_practice['math score'].median()
df_practice['math score'].fillna(median_math_score, inplace=True)

# 4. Fill missing writing scores with the MEAN writing score
mean_writing_score = df_practice['writing score'].mean()
df_practice['writing score'].fillna(mean_writing_score, inplace=True)

# 5. Verify no missing values remain
print("\nMissing values after filling:")
print(df_practice.isnull().sum())
```

```python
#Task 3.2: Basic Groupby Analysis
# Comprehensive groupby analysis
# Your tasks:

# 1. Group by 'gender' and calculate:
#    - Average scores in all subjects
#    - Count of students

# 2. Group by 'test preparation course' and show:
#    - Mean scores for each subject
#    - Show which subject benefits most from test prep

# 3. Group by 'parental level of education' and calculate:
#    - Average math score for each education level
#    - Rank education levels by math performance

# 4. Group by 'lunch' type and show:
#    - Average scores for each subject
#    - Count of students in each lunch category

# my code here:
# 1. Group by 'gender' and calculate average scores and count
print("Analysis by Gender:")
gender_group = df.groupby('gender')[['math score', 'reading score', 'writing score']].agg(['mean', 'count'])
print(gender_group)

# 2. Group by 'test preparation course' and show mean scores
print("\nAnalysis by Test Preparation Course:")
test_prep_group = df.groupby('test preparation course')[['math score', 'reading score', 'writing score']].mean()
print(test_prep_group)
print("\nSubject benefiting most from test prep:")
# Calculate the difference in mean scores between 'completed' and 'none' for each subject
test_prep_diff = test_prep_group.loc['completed'] - test_prep_group.loc['none']
print(test_prep_diff.idxmax(), "benefits most from test prep.")


# 3. Group by 'parental level of education' and calculate and rank average math score
print("\nAnalysis by Parental Level of Education (Math Score):")
parental_edu_math = df.groupby('parental level of education')['math score'].mean().sort_values(ascending=False)
print(parental_edu_math)
print("\nRanking of education levels by math performance:")
print(parental_edu_math.rank(ascending=False))


# 4. Group by 'lunch' type and show average scores and count
print("\nAnalysis by Lunch Type:")
lunch_group = df.groupby('lunch')[['math score', 'reading score', 'writing score']].agg(['mean', 'count'])
print(lunch_group)
```

```python
#Task 3.3: Use your data analysis skills to answer these easy questions:

# 1. Who performs better in math - males or females?
#    - Calculate average math score by gender
#    - Show the difference between male and female average

# 2. Which subject do students perform best in overall?
#    - Calculate the overall average for math, reading, and writing
#    - Rank the subjects from highest to lowest average score

# 3. What's the impact of parental education?
#    - Find the average total score for each parental education level
#    - Identify which education level leads to highest student performance

# my code here:
# 1. Who performs better in math - males or females?
print("1. Who performs better in math - males or females?")
math_avg_by_gender = df.groupby('gender')['math score'].mean()
print("Average math score by gender:\n", math_avg_by_gender)
math_score_difference = math_avg_by_gender['male'] - math_avg_by_gender['female']
print(f"Difference (Male - Female): {math_score_difference:.2f}")
if math_score_difference > 0:
    print("Males perform better in math.")
else:
    print("Females perform better in math.")

# 2. Which subject do students perform best in overall?
print("\n2. Which subject do students perform best in overall?")
overall_avg_scores = df[['math score', 'reading score', 'writing score']].mean().sort_values(ascending=False)
print("Overall average scores for each subject:\n", overall_avg_scores)
print("\nRanking of subjects by overall average score:")
print(overall_avg_scores.rank(ascending=False))
print(f"Students perform best in {overall_avg_scores.index[0]}.")

# 3. What's the impact of parental education?
print("\n3. What's the impact of parental education?")
df['total score'] = df['math score'] + df['reading score'] + df['writing score']
parental_edu_total_avg = df.groupby('parental level of education')['total score'].mean().sort_values(ascending=False)
print("Average total score by parental education level:\n", parental_edu_total_avg)
print(f"\n{parental_edu_total_avg.index[0]} leads to the highest student performance.")
```
