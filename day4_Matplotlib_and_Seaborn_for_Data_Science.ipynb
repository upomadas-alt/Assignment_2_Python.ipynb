# **Gym Data Exploration**

```python
from google.colab import drive
import warnings
warnings.filterwarnings('ignore')
gym = pd.read_csv('/content/drive/MyDrive/Datasets/gym_membership.csv')
```

```python
# Gym data

# Display last 5 rows
display(gym.tail())

# Check data types
display(gym.info())
```

```python
# Task 1: Basic Matplotlib (3 plots)

# 1. Bar chart of gym membership types
# Use: gym['abonoment_type'].value_counts()
membership_counts = gym['abonoment_type'].value_counts()
plt.figure(figsize=(8, 5))
plt.bar(membership_counts.index, membership_counts.values)
plt.title('Distribution of Gym Membership Types')
plt.xlabel('Membership Type')
plt.ylabel('Count')
plt.show()


# 2. Histogram of time in gym
# Use: plt.hist(gym['avg_time_in_gym'])
plt.figure(figsize=(8, 5))
plt.hist(gym['avg_time_in_gym'], bins=20, edgecolor='black')
plt.title('Distribution of Average Time in Gym')
plt.xlabel('Average Time in Gym (minutes)')
plt.ylabel('Frequency')
plt.show()


# 3. Scatter plot: Age vs visits per week
# Use: plt.scatter(gym['Age'], gym['visit_per_week'])
plt.figure(figsize=(8, 5))
plt.scatter(gym['Age'], gym['visit_per_week'], alpha=0.5)
plt.title('Age vs. Visits Per Week')
plt.xlabel('Age')
plt.ylabel('Visits Per Week')
plt.show()
```

```python
# 1. Box plot: Gender vs Age
# Use: sns.boxplot(data=gym, x='gender', y='Age')
plt.figure(figsize=(8, 5))
sns.boxplot(data=gym, x='gender', y='Age')
plt.title('Age Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()

# 2. Count plot: Days per week
# Use: sns.countplot(data=gym, x='days_per_week')
plt.figure(figsize=(10, 6))
sns.countplot(data=gym, x='days_per_week')
plt.title('Count of Visits per Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Count')
plt.show()

# 3. Bar plot: Average visits by membership type
# Use: sns.barplot(data=gym, x='abonoment_type', y='visit_per_week')
plt.figure(figsize=(8, 5))
sns.barplot(data=gym, x='abonoment_type', y='visit_per_week')
plt.title('Average Visits per Week by Membership Type')
plt.xlabel('Membership Type')
plt.ylabel('Average Visits Per Week')
plt.show()
```
